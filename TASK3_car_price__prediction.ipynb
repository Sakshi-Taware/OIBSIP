```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection and Processing

```python
car=pd.read_csv("/content/car data.csv")
```

```python
car.head()
```

```python
# checking the number of rows and columns
car.shape
```

```python
# getting some information about the dataset
car.info()
```

```python
# checking the number of missing values
car.isnull().sum()
```

```python
# checking the distribution of categorical data
print(car.Fuel_Type.value_counts())
print(car.Seller_Type.value_counts())
print(car.Transmission.value_counts())
```

Encoding the Categorical Data

```python
# encoding "Fuel_Type" Column
car.replace({"Fuel_Type":{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
# encoding "Seller_Type" Column
car.replace({"Seller_Type":{'Dealer':0 ,'Individual':1}},inplace=True)
# encoding "Transmission" Column
car.replace({"Transmission":{'Manual':0,'Automatic':1}},inplace=True)

car
```

```python
# Let's get some information about our target feature
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(car["Selling_Price"])

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=car["Selling_Price"])

plt.show()
```

Splitting the data and Target

```python
X = car.drop(['Car_Name','Selling_Price'],axis=1)
Y = car['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
# heatmap of the data for checking the correlation between the numerical features and target column.
numerical_feature = [feature for feature in car.columns if car[feature].dtypes not in ['O', 'object'] ]
plt.figure(figsize = (18, 7))
sns.heatmap(car[numerical_feature].corr(), annot = True, fmt = '0.2f', annot_kws = {'size' : 15}, linewidth = 2, linecolor = 'orange')
plt.show()
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

Model Training

1. Linear Regression

```python
# loading the linear regression model
lin_reg_model= LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

lin_reg_model.fit(X_train,Y_train)

Model Evaluation

```python
# prediction on Training data
prediction= lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_src = metrics.r2_score(Y_train,prediction)
print('ERROR SCORE:',error_src)
```

Visualize the actual prices and Predicted prices

```python
plt.scatter(Y_train,prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices using linear regression model')
plt.show()
```

```python
# prediction on Training data
prediction= lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_src = metrics.r2_score(Y_test,prediction)
print('ERROR SCORE:',error_src)
```

```python
plt.scatter(Y_test,prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
